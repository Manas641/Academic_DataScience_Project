```python
lc = pd.read_csv("LandClass.csv")
```

```python
lc['wildfire_area_mean'] = [reg_nsw['Estimated_fire_area'].mean(),reg_nt['Estimated_fire_area'].mean(),
         reg_ql['Estimated_fire_area'].mean(),reg_sa['Estimated_fire_area'].mean(),
         reg_ta['Estimated_fire_area'].mean(),reg_wa['Estimated_fire_area'].mean(),
         reg_vi['Estimated_fire_area'].mean()]
```

```python
lc['Area'] = [801600,1346200,1727200, 984000,67800,2525500, 227600] 
lc['Permanet_waterbodies_Area'] = (lc['Permanent water bodies']*lc['Area'])/100
lc['open_evergreen_Area'] = (lc['Open forest, evergreen broad leaf']*lc['Area'])/100
lc['closed_evergreen_Area'] = (lc['Closed forest, evergreen, broad leaf']*lc['Area'])/100
lc['closed_deciduous_Area'] = (lc['Closed forest, deciduous broad leaf']*lc['Area'])/100
lc['open_deciduous_Area'] = (lc['Open forest, deciduous broad leaf']*lc['Area'])/100
lc['closed_forest_unknow_Area'] = (lc['Closed forest, unknown']*lc['Area'])/100
lc['open_forest_unknown_Area'] = (lc['Open forest, unknown definitions']*lc['Area'])/100
lc['Shrubs_Area'] = (lc['Shrubs']*lc['Area'])/100
lc['Herbaceous_wetland_Area'] = (lc['Herbaceous wetland']*lc['Area'])/100
lc['Herbaceous_vegetation_Area'] = (lc['Herbaceous vegetation']*lc['Area'])/100
lc['open_sea_area'] = (lc['Open sea']*lc['Area'])/100
lc['Urban_area'] = (lc['Urban / built up']*lc['Area'])/100
lc['bareland_area'] = (lc['Bare / sparse vegetation']*lc['Area'])/100
lc['cropland_area'] = (lc['Cultivated and managed vegetation/agriculture (cropland)']*lc['Area'])/100
```

```python
fig,ax = plt.subplots()
ax.scatter(lc['Shrubs_Area'],lc['wildfire_area_mean'])
ann = ['NSW', 'NT', 'QL', 'SA', 'TA', 'WA', 'VI']
for i in range(7):
    ax.annotate(ann[i],(lc['Shrubs_Area'][i],lc['wildfire_area_mean'][i]))
plt.savefig("shrubs_area.png")
```

```python
fig,ax = plt.subplots()
ax.scatter(lc['Dec_Area'],lc['wildfire_area_mean'])
ann = ['NSW', 'NT', 'QL', 'SA', 'TA', 'WA', 'VI']
for i in range(7):
    ax.annotate(ann[i],(lc['Dec_Area'][i],lc['wildfire_area_mean'][i]))
plt.savefig("deciduos_area.png")
```

```python
fig,ax = plt.subplots()
ax.scatter(lc['Area'],lc['wildfire_area_mean'])
ann = ['NSW', 'NT', 'QL', 'SA', 'TA', 'WA', 'VI']
for i in range(7):
    ax.annotate(ann[i],(lc['Area'][i],lc['wildfire_area_mean'][i]))
ax.set_xlabel("Area of region",fontsize='large')
ax.set_ylabel("Mean area of wildfire",fontsize='large')
plt.savefig("area_area.png")
```
