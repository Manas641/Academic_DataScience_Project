```python
X = reg_nt[['Precipitation_max', 'Precipitation_mean',
       'RelativeHumidity_max', 'RelativeHumidity_mean', 'SoilWaterContent_max',
       'SoilWaterContent_mean', 'SolarRadiation_max', 'SolarRadiation_mean',
       'Temperature_min', 'Temperature_max', 'Temperature_mean',
       'WindSpeed_min', 'WindSpeed_max', 'WindSpeed_mean',
       'Vegetation_index_mean']]

Y = reg_nt['Mean_estimated_fire_brightness']

X_test = reg_wa[['Precipitation_max', 'Precipitation_mean',
       'RelativeHumidity_max', 'RelativeHumidity_mean', 'SoilWaterContent_max',
       'SoilWaterContent_mean', 'SolarRadiation_max', 'SolarRadiation_mean',
       'Temperature_min', 'Temperature_max', 'Temperature_mean',
       'WindSpeed_min', 'WindSpeed_max', 'WindSpeed_mean',
       'Vegetation_index_mean']]

Y_test = reg_wa['Mean_estimated_fire_brightness']
```

```python
from sklearn.metrics import mean_squared_error, make_scorer

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X,Y)
pred1 = lr.predict(X_test)
np.sqrt(mean_squared_error(Y_test,pred1))
```
